```python
import numpy as np
```

```python
X = np.arange(-20,20, 0.2)
Y = X**2 - 6*X + 8
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X,Y)
```

```python
def my_pol(x):
    return x**2 -6*x+8
```

```python
cur_x = -15

def draw(x):
    plt.plot(X,Y)
    plt.scatter(x, my_pol(x), c = 'r')
    plt.show()
```

```python
draw(cur_x)
```

```python
def der(x):
    return 2*x - 6
```

```python
x_new = cur_x - der(cur_x)
x_new
```

```python
cur_x = x_new
x_new = cur_x - der(cur_x)
x_new
```

```python
lr = 0.01

for i in range(100):
    cur_x = x_new
    x_new = cur_x -lr* der(cur_x)
    x_new
draw(cur_x)
print (cur_x)
```
